```python
#function to add targets
def add_target_ret_time(
    df: pl.DataFrame,
    delta: int = 50,  # in milliseconds
    base_col: str = "mid",  # 'mid' or 'vwap'
    levels: int = 1,
    depth: int = 1,
    time_col: str = "ts_dt",
    products: list[str] | None = None
) -> pl.DataFrame:

    if time_col not in df.columns:
        df = add_ts_dt(df)

    # Ensure time_col is datetime[ns]
    if df[time_col].dtype != pl.Datetime("ns"):
        df = df.with_columns(pl.col(time_col).cast(pl.Datetime("ns")))

    # Ensure price column exists
    if base_col == "mid":
        col_prefix = "mid"
        if col_prefix not in df.columns:
            df = add_mid(df, products=products)
    elif base_col == "vwap":
        col_prefix = f"vwap_level{levels}"
        if col_prefix not in df.columns:
            df = add_vwap(df, levels=levels, depth=depth, products=products)
    else:
        raise ValueError("base_col must be 'mid' or 'vwap'.")

    

    price_col = col_prefix
    future_df = (
        df.select([time_col, price_col])
        .with_columns(
            (pl.col(time_col) + pl.duration(milliseconds=-delta))
            .cast(pl.Datetime("ns"))
            .alias(time_col)
        )
        .rename({price_col: f"{price_col}_fut"})
    )
    ret_col_name = (
            f"ret_{delta}ms_fut"
        )
    df = df.join_asof(
        future_df,
        left_on=time_col,
        right_on=time_col,
        strategy="backward",
        tolerance=timedelta(milliseconds=1000000)  # generous tolerance
    ).with_columns([
        (pl.col(f"{price_col}_fut")/ (pl.col(price_col) ) - 1).alias(ret_col_name)
    ]).drop([f"{price_col}_fut"])


    return df.drop_nulls()
```

```python


def correlation_xgb_feature_selection(
    df_pl: pl.DataFrame,
    features: list,
    targets: list,
    num_feat: list,
    corr_threshold: float = 0.98,
    n_splits: int = 5,
    min_fold_count: int = 3
):
    """
    Select features using correlation filtering + XGBoost.
    This version includes robustness to feature/target outliers.
    NOTE: Uses KFold as per user specification, which may introduce look-ahead bias.
    """
    # Drop all-zero columns
    cols_to_drop = [col for col in features if (df_pl[col] == 0).all()]
    df_pl = df_pl.drop(cols_to_drop)
    reduced_features_0 = [f for f in features if f not in cols_to_drop]

    # --- Step 1: Robust Correlation filtering using Spearman ---
    # corr_matrix = df_pl.select(reduced_features_0).to_pandas().corr(method='spearman').abs()
    corr_matrix = df_pl.sample(n=2_000_000, shuffle=True, seed=42) \
                   .select(reduced_features_0) \
                   .to_pandas() \
                   .corr(method='spearman').abs()
    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [col for col in upper.columns if any(upper[col] > corr_threshold)]
    reduced_features = [f for f in reduced_features_0 if f not in to_drop]
    
    print(f'Dropped {len(to_drop)} features via correlation.')
    
    # --- Step 2: XGBoost with stability selection ---
    all_importances = []
    stable_features_all_targets = set()

    for target, top_n in zip(targets, num_feat):
        print(f"Processing Target: {target}")
        X = df_pl.select(reduced_features).to_pandas()
        
        # <--- CORRECTED SECTION STARTS HERE ---
        y_raw = df_pl[target]
        # First, calculate the quantile values from the Polars Series
        lower_bound = y_raw.quantile(0.005)
        upper_bound = y_raw.quantile(0.995)
        # Then, use the standard .clip() method with these calculated bounds
        y = y_raw.clip(lower_bound=lower_bound, upper_bound=upper_bound).to_pandas()
        # <--- CORRECTED SECTION ENDS HERE ---

        # Keeping KFold as requested
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
        fold_importances = []
        feature_counts = {f: 0 for f in reduced_features}

        for train_idx, val_idx in kf.split(X):
            X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
            
            # Use a robust loss function (Huber Loss) in XGBoost
            model = XGBRegressor(
                # objective='reg:pseudohubererror',
                n_estimators=300,
                max_depth=5,
                learning_rate=0.05,
                random_state=42,
                n_jobs=-1
            )
            model.fit(X_train, y_train)
            importances = model.feature_importances_

            # ... (rest of the function is the same)
            top_features_fold = [f for f, _ in sorted(zip(reduced_features, importances), key=lambda x: x[1], reverse=True)[:top_n]]
            for f in top_features_fold:
                feature_counts[f] += 1
            fold_importances.append(importances)

        stable_features = [f for f, count in feature_counts.items() if count >= min_fold_count]
        stable_features_all_targets.update(stable_features)
        avg_importances = np.mean(fold_importances, axis=0)
        imp_df = pd.DataFrame({"feature": reduced_features, "importance": avg_importances, "target": target}).sort_values("importance", ascending=False)
        all_importances.append(imp_df)
        print(stable_features)
    # --- Step 3: Combine importances across targets ---
    all_importances_df = pd.concat(all_importances)
    avg_importance = (
        all_importances_df.groupby("feature")["importance"]
        .mean()
        .sort_values(ascending=False)
        .reset_index()
    )
    final_features = [f for f in avg_importance["feature"].tolist() if f in stable_features_all_targets]

    return final_features, avg_importance
```

```python
def correlation_xgb_feature_selection_lazy(
    df_lazy: pl.LazyFrame, # <-- Accepts a LazyFrame
    features: list,
    targets: list,
    num_feat: list,
    corr_threshold: float = 0.90, # <-- Using the 0.90 threshold
    n_splits: int = 5,
    min_fold_count: int = 3
):
    """
    Performs feature selection on a LazyFrame to handle larger-than-RAM data.
    """
    # Drop all-zero columns (this can be done lazily)
    # Note: This check still requires a full scan, but it's a one-off cost.
    print("Checking for all-zero columns...")
    all_zeros_check = df_lazy.select([(pl.col(c) == 0).all() for c in features]).collect()
    cols_to_drop = [features[i] for i, col in enumerate(all_zeros_check.iter_cols()) if col[0]]
    
    df_lazy = df_lazy.drop(cols_to_drop)
    reduced_features_0 = [f for f in features if f not in cols_to_drop]

    # --- Step 1: Correlation filtering on a subsample (done lazily) ---
    print("Calculating correlation matrix on a 2M row subsample...")
    # We can sample the LazyFrame and only .collect() the small sample
    corr_matrix = df_lazy.sample(n=2_000_000, shuffle=True, seed=42) \
                         .select(reduced_features_0) \
                         .collect() \
                         .to_pandas() \
                         .corr(method='spearman').abs()

    upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
    to_drop = [col for col in upper.columns if any(upper[col] > corr_threshold)]
    reduced_features = [f for f in reduced_features_0 if f not in to_drop]
    
    print(f'Dropped {len(to_drop)} features via correlation.')
    
    # Now, create a new LazyFrame with only the surviving features and targets
    final_feature_set = reduced_features + targets
    df_lazy_reduced = df_lazy.select(final_feature_set)

    # --- Step 2: XGBoost with stability selection ---
    all_importances = []
    stable_features_all_targets = set()

    for target, top_n in zip(targets, num_feat):
        print(f"Processing Target: {target}")
        
        # --- LAZY-AWARE DATA PREPARATION ---
        # Select only necessary columns for this target
        y_lazy = df_lazy_reduced.select(pl.col(target))
        X_lazy = df_lazy_reduced.select(reduced_features)

        # Perform clipping lazily and collect ONLY the target series 'y'
        print("  - Clipping target variable...")
        y = y_lazy.with_columns(
            pl.col(target).clip(
                lower=pl.col(target).quantile(0.005),
                upper=pl.col(target).quantile(0.995)
            )
        ).collect().to_pandas().squeeze()

        # IMPORTANT: Collect the features 'X' right before the loop
        # This is the main memory load, but now only on the reduced feature set
        print("  - Collecting features for model training...")
        X = X_lazy.collect().to_pandas()

        kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)
        # ... (The rest of the XGBoost loop is the same as before) ...
        fold_importances = []
        feature_counts = {f: 0 for f in reduced_features}

        for train_idx, val_idx in kf.split(X):
            X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
            model = XGBRegressor(
                objective='reg:pseudohubererror', # Re-enable robust loss
                n_estimators=300, max_depth=5, learning_rate=0.05,
                random_state=42, n_jobs=-1
            )
            model.fit(X_train, y_train)
            importances = model.feature_importances_
            top_features_fold = [f for f, _ in sorted(zip(reduced_features, importances), key=lambda x: x[1], reverse=True)[:top_n]]
            for f in top_features_fold:
                feature_counts[f] += 1
            fold_importances.append(importances)

        stable_features = [f for f, count in feature_counts.items() if count >= min_fold_count]
        stable_features_all_targets.update(stable_features)
        avg_importances = np.mean(fold_importances, axis=0)
        imp_df = pd.DataFrame({"feature": reduced_features, "importance": avg_importances, "target": target}).sort_values("importance", ascending=False)
        all_importances.append(imp_df)

    # ... (Final combination step is the same) ...
    all_importances_df = pd.concat(all_importances)
    avg_importance = (
        all_importances_df.groupby("feature")["importance"]
        .mean().sort_values(ascending=False).reset_index()
    )
    final_features = [f for f in avg_importance["feature"].tolist() if f in stable_features_all_targets]
    return final_features, avg_importance
```
